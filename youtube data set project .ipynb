```python
# 12. Find the title of the video which has the maximum number of dislikes and the title of the video having minimum dislikes and write your inferences. 
```

```python
#video with the maximum number of likes
max+likes_video = df[df['likes'] == df['likes'].max()]

#video with the minimum number of likes
minlikes_video = df[df['likes'] == df['likes'].min()]

#videos with max and min likes
title_max_likes = max_likes_video['title'].values[0]
title_min_likes = min_likes_video['title'].values[0]

# Print the titles
print("Title of the video with the maximum number of likes:", title_max_likes)
print("Title of the video with the minimum number of likes:", title_min_likes)
```

```python
# 13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.
```

```python
correlation = df['view_count'].corr(df['dislikes'])
correlation
```

```python
df[['dislikes']].isnull().sum(),
```

```python
df[['view_count']].isnull().sum(),
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting
sns.set(style="darkgrid")
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='view_count', y='dislikes', alpha=0.7)
plt.title('Relationship between View Count and Dislikes')
plt.text(df['view_count'].max() * 0.8, df['dislikes'].max() * 0.9, f'Correlation: {correlation:.2f}', fontsize=12, color='red')
plt.show()
print(f"Correlation between view count and dislikes: {correlation:.2f}")
```
